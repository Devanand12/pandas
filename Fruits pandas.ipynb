```python
import pandas as pd
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/yunglinchang/DataScience_playground/master/Pandas/data/TheFruit_sales.csv')
```

```python
df.head()
```

### Sales_ID: Weird Exclamation Mark and Redundant Spaces

```python
df['Sales_ID']
```

```python
df['Sales_ID']=df['Sales_ID'].str.rstrip('!').str.strip()
```

```python
df.head()
```

### Sales_Branch: Missing?

```python
df['Sales_ID'].str[:2]
```

```python
df.Sales_Branch=df['Sales_ID'].str.slice(0,2)
```

```python
df.head()
```

### Product_Description: Long Descriptions containing which Products are Sold

```python
df.Product_Description[0]
```

```python
df.Product_Description.str.split(" Product:")
```

```python
df.Product_Description.str.split(" Product:",expand=True)
```

```python
df.Product_Description.str.split(" Product:",expand=True)[1]
```

```python
df[['Description','Products']]=df.Product_Description.str.split(" Product:",expand=True)
```

```python
df.head()
```

```python
df.Products=df.Products.str.split().apply(sorted)
```

```python
df.head()
```

### Product_Count: The Number of Fruits for each Salesperson

```python
df.Product_Count=df.Products.apply(len)
```

```python
df.Product_Count
```

###  Who sells the most variety of fruits in The Fruit!

```python
df[df.Product_Count==max(df.Product_Count)]
```
